__`UCRCOUNT`__: Coded as 1 or -1. 1 indicates a crime that was reported during the given month, and the -1 refers to any crime that was subtracted or unfounded from a previous reporting period. 

`COMPLAINTYEAR`: Year of the complaint

__`COMPLAINTNUM`__: departmental complaint number

__`UCR_OFFENSE`: string of the UCR crime__

`UCR_CRIME_CODE`: number code that corresponds to the UCR offense

`TYPE`: __Person__=homicide, robbery, aggravated assault, and rape. __Property__=burglary, larceny, and motor vehicle theft

`MONTH`: The month that the crime data was submitted to the State of Missouri.

`YEAR`: Year data was submitted to the State of Missouri

__`DT_CALLREC`: Date and time that the call for the crime was received__

__`D_OCCURRED`: date that the crime occurred__

`DOW_OCCURRED`: day of the week that the crime occurred

`ZONE`: geographical zone where the crime occurred

__`ADDRESS`: address of the crime, excluded for rapes and human trafficking cases__

`PRECINCT`: The precinct where the crime occurred. For St. Louis County PD: North County Precinct, Central County Precinct, Affton Southwest Precinct, South County Precinct, City of Fenton Precinct, City of Wildwood Precinct, West County Precinct, City of Jennings Precinct, MetroLink Police Unit. For Ellisville PD: East and West.PREMISE: The premise of the crime (eg. residential, business, etc.)

`REPORTING_JURIS`: The jurisdiction that reported this data to the State of Missouri (Saint Louis County Parks data is reported by Saint Louis County).

`FOR_JURIS`: The jurisdiction the data is for. This will either be a municipality name, MetroLink, or Saint Louis County.

__`X`: Longitude. If UCRCOUNT=-1 this field is blank. Additionally, if address is redacted this field will be blank.__

__`Y`: Latitude: If UCRCOUNT=-1 this field is blank. Additionally, if address is redacted this field will be blank.__

### Visualization of Missing Values

```python
import matplotlib.style as style #sets the size of the charts
plt.figure(figsize=(10,6))
plt.title('Missing Values (HeatMap) - YEAR 2017', fontsize = 16)
sns.heatmap(df_2017.isnull(),cbar=False,yticklabels=False,cmap = 'gnuplot2') #Check 'cmap'
plt.show()
```

```python
import matplotlib.style as style #sets the size of the charts
plt.figure(figsize=(10,6))
plt.title('Missing Values (HeatMap) - YEAR 2018', fontsize = 16)
sns.heatmap(df_2018.isnull(),cbar=False,yticklabels=False,cmap = 'copper_r') # Check 'cmap'
plt.show()
```

- __`Missing Values`__ Few of the varibales have missing valies both in the the Year 2017 and 2018. 

- I have removed all the columns with missing values after considering their importance in the data set. 

- Complaints `COMPLAINTYEAR` which are registered only in the year 2017 or 2018 are kep in the respective data set 

- To clean the data, 'ADDRESS', 'PRECINCT', 'X', 'Y', 'UCRCOUNT', 'COMPLAINTNUM', 'DT_CALLREC',
    'REPORTING_JURIS', 'OBJECTID', 'D_OCCURRED', 'COMPLAINTYEAR' have been removed from the data set 

```python
df_2017['COMPLAINTYEAR'].value_counts() 
```

```python
df_2018['COMPLAINTYEAR'].value_counts()
```

```python
df_17 = df_2017[df_2017['COMPLAINTYEAR'] > 16]
print(df_17['COMPLAINTYEAR'].value_counts())
```

```python
df_18 = df_2018[df_2018['COMPLAINTYEAR'] > 17]
print(df_18['COMPLAINTYEAR'].value_counts())
```

```python
### make a clean data frmae after removing the unwanted columns
data_17 = df_17.drop([
    'ADDRESS', 'PRECINCT', 'X', 'Y', 'UCRCOUNT', 'COMPLAINTNUM', 'DT_CALLREC',
    'REPORTING_JURIS', 'OBJECTID', 'D_OCCURRED', 'COMPLAINTYEAR'
],
                     axis=1)
```

```python
data_18 = df_18.drop([
    'ADDRESS', 'PRECINCT', 'X', 'Y', 'UCRCOUNT', 'COMPLAINTNUM', 'DT_CALLREC',
    'REPORTING_JURIS', 'OBJECTID', 'D_OCCURRED', 'COMPLAINTYEAR'
],
                     axis=1)
```

__Cleaned Data from both the years was merged using the `concat` function.__

```python
# Combine both the data in single dataframe
data = pd.concat ([data_17, data_18])
```

## Visualization

__Top10 `For Juris` for the Year 2017 and 2018__

__As they were not common, made plots for each year seprately as well__

__Used Different color pallets 

```python
import matplotlib.style as style
style.use('seaborn-poster') #sets the size of the charts
plt.figure(figsize=(8, 5))
sns.countplot(y=data['FOR_JURIS'],
              hue="YEAR",
              data=data,
              order=data['FOR_JURIS'].value_counts().iloc[:10].index, # totp 10
              palette="rocket" )
```

```python
# Top 10 for Year 2017
plt.figure(figsize=(8, 5))
sns.countplot(y=data_17['FOR_JURIS'],
              hue="YEAR",
              data=data_17,
              order=data_17['FOR_JURIS'].value_counts().iloc[:10].index,
              palette="rocket" ) 
```

```python
# Top 10 for Year 2018
plt.figure(figsize=(8, 5))
sns.countplot(y=data_18['FOR_JURIS'],
              hue="YEAR",
              data=data_18,
              order=data_18['FOR_JURIS'].value_counts().iloc[:10].index,
              palette="Set2" )
```

```python
#Plots of crime `TYPE` in year 2017 and 2018
plt.figure(figsize=(7, 4))
sns.countplot(data=data,
              y='TYPE',
              hue="YEAR",
              order=data['TYPE'].value_counts().index,
              palette="hsv")
plt.plot()
```

```python
#Top 10 crime `Premises`in year 2017 and 2018
plt.figure(figsize=(8, 5))
sns.countplot(y=data['PREMISE'],
              hue="YEAR",
              data=data,
              order=data['PREMISE'].value_counts().iloc[:10].index,
              palette="autumn" )
```

```python
# Number of crimes in different months in the Year 2017 and 2018
plt.figure(figsize=(10, 6))
sns.countplot(data=data,
              y='MONTH',
              hue="YEAR",
              order=data['MONTH'].value_counts().index,
              palette= "YlGnBu")
```

```python
# Plot showing number of crimes in different days of the week in the year 2017 and 2018.  
plt.figure(figsize=(10,6))
sns.countplot(data=data,y='DOW_OCCURRED', hue = "YEAR", order = data['DOW_OCCURRED'].value_counts().index, palette="prism")
plt.plot()
```
